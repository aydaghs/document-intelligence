```python
# 2) Install ngrok for public URL
!pip install pyngrok
from pyngrok import ngrok

# ngrok now requires an auth token (free account)
# 1) Sign up at https://dashboard.ngrok.com/signup
# 2) Copy your authtoken from https://dashboard.ngrok.com/get-started/your-authtoken
# 3) Paste it below and rerun this cell.

NGROK_AUTH_TOKEN = "<YOUR_NGROK_AUTHTOKEN>"
if NGROK_AUTH_TOKEN:
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# 3) Create tunnel and print URL
try:
    tunnel = ngrok.connect(8501, "http")
    print('Open this URL in a browser: ', tunnel.public_url)
except Exception as e:
    print('Error creating ngrok tunnel:', e)

# 4) Run Streamlit (this cell will keep running)
!streamlit run app.py &
```
